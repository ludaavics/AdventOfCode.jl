```julia
function parse_claim(claim_string)
    claim_regex = r"^#(?<id>\d+) @ (?<x>\d+),(?<y>\d+): (?<w>\d+)x(?<h>\d+)"
    claim = [parse(Int, x) for x in  match(claim_regex, claim_string).captures]
end
claims = parse_claim.(split(input, "\n"))
```

```julia
cells_claimed = [(i, j) for (id, x, y, w, h)=claims[1:end] for i=x+1:x+w for j=y+1:y+h]
```

```julia
claim_count = Dict{Tuple, Int}()
for cell in cells_claimed
    global claim_count
    current_val = get(claim_count, cell, 0)
    claim_count[cell] = current_val + 1
end
```

```julia
@time num_duplicates = length([1 for x in values(claim_count) if x > 1])
```

```julia
for claim in claims
    (id, x, y, w, h)=claim
    wants = [(i, j) for i=x+1:x+w for j=y+1:y+h]
    if all([claim_count[c] == 1 for c in wants])
        println("Claim $id is unconstested")
     end
end 
```

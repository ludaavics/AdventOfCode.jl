```julia
parse_input(input) = parse.(Int, split(input))
```

```julia
struct Node
    name::Char
    num_children::Int
    num_metadata::Int
    parent::Union{Node, Nothing}
    children::Array{Node}
    metadata::Array{Int}
end

Node(name, nc, nm, parent) = Node(name, nc, nm, parent, Array{Node}(undef, 0), Array{Node}(undef, 0))
```

```julia
function needs_children(node::Node; debug=false)
    needs = node.num_children
    has = length(node.children)
    if needs > has
        !debug ? nothing : println("Node $(node.name) needs $needs children but only has $has.")
        return true
    elseif needs == has
        !debug ? nothing : println("Node $(node.name) has all $needs children it needs.")
        return false
    else
        throw("Node $(node.name) has $has children but it only needs $needs!")
    end
end
```

```julia
function build_tree(input; debug=false)
    input = parse_input(input)
    metadata = Int[]
    
    queue = Array{Node}(undef, 0)
    name = 'A'
    root = Node(name, input[1:2]..., nothing)
    push!(queue, root)
    name += 1

    cursor = 3
    N = length(input)
    while cursor < N
        parent = queue[end]
        if needs_children(parent, debug=debug)
            child = Node(name, input[cursor:cursor + 1]..., parent)
            push!(queue, child)
            push!(parent.children, child)
            name += 1
            cursor += 2
        else
            data = input[cursor:cursor + parent.num_metadata - 1]
            append!(parent.metadata, data)
            append!(metadata, data)
            pop!(queue)
            cursor += parent.num_metadata
        end
    end
    return (root, metadata)
end
```

```julia
@time begin
    (root, metadata) = build_tree(input)
    sum(metadata)
end
```

```julia
function value(node::Node; debug=false)
    if node.num_children == 0
        val = sum(node.metadata)
        !debug ? nothing : println("Node $(node.name) has no children and value $val")
        return val
    else
        val = sum([node.metadata[i] <= node.num_children ? 
                   value(node.children[node.metadata[i]]) : 
                   0 for i=1:node.num_metadata])
    end
    
end
```

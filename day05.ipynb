```julia
switch_case(c) = isuppercase(c) ? lowercase(c) : uppercase(c)
```

```julia
function react(polymer)
    
    checked = 1
    offsets = []
    while true
        m = match(r"([A-Z])\1"i, polymer, checked)
        if m === nothing
            break
        end
        push!(offsets, m.offsets[1])
        checked = m.offsets[1] + 1
    end
    
    if length(offsets) == 0
        return polymer
    end

    for offset in reverse(offsets)
        if offset == length(polymer)
            # 3 times the same letter at the enf of the string, 
            # and the last two reacted
            continue
        end
        
        if polymer[offset + 1] == switch_case(polymer[offset]) 
            if offset == 1
                polymer = polymer[3:end] 
            else
                polymer = polymer[1:offset - 1] * polymer[offset + 2:end]
            end
        end
    end
    
    return polymer
end
```

```julia
function breakdown(polymer)
    while true
        new_polymer = react(polymer)
        if new_polymer == polymer
            return polymer
        else
            polymer = new_polymer
        end
    end
end
```

```julia
units(polymer) = unique(lowercase(polymer))
```

```julia
remove_unit(polymer, unit) = replace(polymer, Regex("($unit)", "i") => s"")
```

```julia
function smallest_polymer_loop(polymer)
    min_unit = ""
    min_length = length(polymer)
    
    for unit in units(polymer)
        sub_polymer = breakdown(remove_unit(polymer, unit))
        sub_polymer_length = length(sub_polymer)
        min_unit, min_length = sub_polymer_length < min_length ? (unit, sub_polymer_length) : (min_unit, min_length)
    end
    return (min_length, min_unit)
end
```

```julia
function smallest_polymer_comprehensions(polymer)
    _units = units(polymer)
    min_length, min_index = findmin([length(breakdown(remove_unit(polymer, unit))) for unit in _units])
    min_unit =  _units[min_index]
    return (min_length, min_unit)
end
```

```julia
smallest_polymer_comprehensions(test_input)
```

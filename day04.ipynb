```julia
function parse_events(event_string)
    event_regex = raw"^\[(?<year>\d{4})\-(?<month>\d{2})\-(?<day>\d{2}) (?<hour>\d{2}):(?<minute>\d{2})\] "
    event_regex *= raw"(Guard #(?<id>\d+) (?<action_>.*)|(?<action>.*))"
    event_regex = Regex(event_regex)
    m = match(event_regex, event_string)
    event = [
        join(m.captures[1:3], "-") * " " * join(m.captures[4:5], ":"),
        parse(Int, m[:year]),
        parse(Int, m[:month]),
        parse(Int, m[:day]),
        parse(Int, m[:hour]),
        parse(Int, m[:minute]),
        m[:id] !== nothing ? parse(Int, m[:id]) : m[:id],
        m[:id] !== nothing ? m[:action_] : m[:action]
    ]
end
events = sort(
    parse_events.(split(input, "\n")),
    lt=(x, y) -> isless(x[1], y[1])
)
for (i, event) in enumerate(events)
    event[7] = event[7] === nothing ? events[i - 1][7] : event[7]
end
events
```

```julia
schedule = Dict{Tuple{String, Int}, Array{Bool, 1}}()
for event in events
    date = event[1][1:10]
    minute = event[6]
    guard_id = event[7]
    action = event[end]
    if action == "begins shift"
        continue
    end
    sleeping = get(schedule, (date, guard_id), falses(60))
    if action == "falls asleep"
        sleeping[minute + 1:end] = fill(true, 60 - minute)
    elseif action == "wakes up"
        sleeping[minute + 1:end] = fill(false, 60 - minute)
    end
    schedule[(date, guard_id)] = sleeping
end
schedule
```

```julia
sleep_time_by_guard = Dict{Int, Int}()
for ((date, guard_id), sleeping) in schedule
    sleep_time_by_guard[guard_id] = get(sleep_time_by_guard, guard_id, 0) + sum(sleeping)
end
sleep_time_by_guard
```

```julia
sleepiest_guard = reduce(
    (x, y) -> sleep_time_by_guard[x] >= sleep_time_by_guard[y] ? x : y, 
    keys(sleep_time_by_guard)
)
```

```julia
sleepiest_schedule = filter(x -> x[1][2] == sleepiest_guard, schedule)
sleepiest_minute = argmax(sum(values(sleepiest_schedule))) - 1
```

```julia
sleepiest_guard * sleepiest_minute
```

```julia
sleep_proba_schedule_by_guard = Dict{Int, Tuple{Array{Int}, Array{Float64}}}()
for ((date, guard_id), sleeping) in schedule
    N, current_avg = get(
        sleep_proba_schedule_by_guard,
        guard_id,
        (ones(60), zeros(60))
    )
    new_avg = current_avg + (sleeping - current_avg) ./ N
    new_avg = [isnan(x) ? 0 : x for x in new_avg]
    N .+= 1
    sleep_proba_schedule_by_guard[guard_id] = (N, new_avg)
    if guard_id == 113
         println("$(sleeping[1:5]) -> $(new_avg[1:5]) and count $(N[1:5])")
    end
end
sleep_proba_schedule_by_guard
```

```julia
guard_with_sleepiest_minute = reduce(
    (x, y) -> maximum(sleep_proba_schedule_by_guard[x][2]) >= maximum(sleep_proba_schedule_by_guard[y][2]) ? x : y, 
    keys(sleep_time_by_guard)
)
```
